```python
from environment import UAVMissionEnv
import numpy as np

def test_always_continue():
    print("=== 测试 1: 始终选择‘继续任务’ (对应论文的 Policy 3: Never abort) ===")
    env = UAVMissionEnv()
    # 设定固定的随机种子以便观察复现
    state, _ = env.reset(seed=42) 
    done = False
    step = 0
    total_reward = 0
    
    print(f"初始状态: {state}")
    while not done:
        # 动作 1 代表 DC (继续任务)
        action = 1
        next_state, reward, done, _, info = env.step(action)
        total_reward += reward
        step += 1
        
        # 打印每个时间步的退化状态，E11(相机), E21(齿轮箱组件1), E22(齿轮箱组件2), n(时间步)
        print(f"Step {step:2d} | 动作: 继续 (DC) | "
              f"状态 (E11, E21, E22, n): [{next_state:.3f}, {next_state[8]:.3f}, {next_state[9]:.3f}, {next_state[10]:.0f}] | "
              f"单步奖励: {reward:.1f}")
        
        if done:
            print(f"-> 回合结束! 总奖励: {total_reward:.1f} | 结束原因: {info.get('reason', 'Unknown')}\n")

def test_abort_midway():
    print("=== 测试 2: 中途选择‘中止任务’ (验证救援阶段和成本结算) ===")
    env = UAVMissionEnv()
    state, _ = env.reset(seed=10)
    done = False
    step = 0
    total_reward = 0
    
    print(f"初始状态: {state}")
    while not done:
        # 在第 5 步时强制选择中止任务 (动作 0 代表 DA)
        action = 0 if step == 5 else 1
        action_name = "中止 (DA)" if action == 0 else "继续 (DC)"
        
        next_state, reward, done, _, info = env.step(action)
        total_reward += reward
        step += 1
        
        print(f"Step {step:2d} | 动作: {action_name} | "
              f"状态 (E11, E21, E22, n): [{next_state:.3f}, {next_state[8]:.3f}, {next_state[9]:.3f}, {next_state[10]:.0f}] | "
              f"单步奖励: {reward:.1f}")
        
        if done:
            print(f"-> 回合结束! 总奖励: {total_reward:.1f} | 结束原因: {info.get('reason', 'Unknown')}\n")

test_always_continue()
  
```
